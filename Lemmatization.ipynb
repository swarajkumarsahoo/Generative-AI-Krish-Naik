Lemmatization technique is like stemming. The output we will get after lematization is called 'lemma' rather than root stem, the output of stemming. After lemmatization we would be getting a valid word that means the same thing.

NLTK provides WordNetLematizer class which is a thin wrapper around the wordnet corpus. This class uses morphy() function to the WordNet CorpusReader class to find a lemma. Example--

from nlkt.stem

```python
from nltk.stem import WordNetLemmatizer
import nltk as nltk
```

```python
lemmatize=nltk.WordNetLemmatizer()
lemmatizer=WordNetLemmatizer()
```

```python
lemmatizer.lemmatize('going')
```

```python
words=['spinning','spin','spins','cluster','clustering','decision','decisioning']
```

```python
for i in words:
    print(i,"----->",lemmatizer.lemmatize(i,pos='v'))
```

Pos parameter used inside lemmatize function

Noun-n
Verb-v
Adjective-a
Adverb-r
