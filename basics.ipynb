List Comprehension

```python
# List Comprehension

# Objective: Create a new list using a exisiting list based on some condition
# Using traditional coding
fruits=["apple","bananna","cherry","kiwi","mango","orange"]
newlist=[]

for x in fruits:
    if "orange" in fruits:
        newlist.append(x)

print(newlist)

# Using List Comprehension

newlist=[x for x in fruits if 'orange' in fruits]

print(newlist)
```

Join Keyword in Python

```python
words =["My","Name","Is","Python"]
result=" ".join(words)
print(result)

# " " is the separator
# words must be string only
```

```python
words =["My","Name","Is","Python"]
for i in range(len(words)):
    words[i]=" ".join(words[i]+"97")

print(words)
```
