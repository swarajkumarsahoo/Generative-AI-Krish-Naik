```python
paragraph="Adolf Hitler was a German political leader who ruled Germany from 1933 to 1945. He led the Nazi Party and established a dictatorship based on extreme nationalism, racism, and authoritarian control. His policies caused World War II and led to the Holocaust, in which millions of Jews and others were systematically murdered. Hitler’s actions resulted in massive destruction, loss of life, and long-lasting global consequences. Today, he is remembered as a warning of how dangerous extremist ideology and unchecked power can be."
```

```python
import nltk
```

```python
# words tokenizer

words=nltk.word_tokenize(paragraph)
```

```python
# Finding points of speech tagging

tag_elements=nltk.pos_tag(words)
```

```python
# Displaying the Named Entity Recognization 

nltk.ne_chunk(tag_elements)
```
