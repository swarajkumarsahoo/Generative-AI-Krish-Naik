Stop Words: Words like have,am, I, Why,......N, doesnt play a big role when performing classification like algorithms 

```python
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')

# downloading languages which are present in the NLTK library₹
```

```python
stopwords.words('english')
```

```python
stemmer=PorterStemmer()
```

```python
paragraph="Adolf Hitler was a German political leader who ruled Germany from 1933 to 1945. He led the Nazi Party and established a dictatorship based on extreme nationalism, racism, and authoritarian control. His policies caused World War II and led to the Holocaust, in which millions of Jews and others were systematically murdered. Hitler’s actions resulted in massive destruction, loss of life, and long-lasting global consequences. Today, he is remembered as a warning of how dangerous extremist ideology and unchecked power can be."
```

```python
sentences=nltk.sent_tokenize(paragraph)

# Take the existing paragraph and divided it into sentences
```

```python
## Apply Stopwords & Filter & Then Apply Stemming
```

```python
for i in range(len(sentences)):

    # Fetching the list of words
    words=nltk.word_tokenize(sentences[i])# list of words
    
    #If a word is present in the stopwords list, then skip/ignore, if not present stem it out using 'stem' function
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))] 

    # Replacing each index of the sentence with stemmed & non stopwords words
    sentences[i]=' '.join(words) #Converting all the list of words into sentences

print(sentences)
```

```python
# using Snowball Stemmer
snowballstemmer=SnowballStemmer('english')
```

```python
for i in range(len(sentences)):

    # Fetching the list of words
    words=nltk.word_tokenize(sentences[i])# list of words
    
    #If a word is present in the stopwords list, then skip/ignore, if not present stem it out using 'stem' function
    words=[snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))] 

    # Replacing each index of the sentence with stemmed & non stopwords words
    sentences[i]=' '.join(words) #Converting all the list of words into sentences

print(sentences)
```

```python
# Using Lemmatization as the results are not good

lemma=WordNetLemmatizer()
```

```python
for i in range(len(sentences)):

    # Fetching the list of words
    words=nltk.word_tokenize(sentences[i])# list of words
    
    #If a word is present in the stopwords list, then skip/ignore, if not present stem it out using 'stem' function
    words=[lemma.lemmatize(word.lower(),pos='v') for word in words if word not in set(stopwords.words('english'))] 

    # Replacing each index of the sentence with stemmed & non stopwords words
    sentences[i]=' '.join(words) #Converting all the list of words into sentences

print(sentences)
```
