```python
corpus_paragraph ="My name is Swaraj Kumar Sahoo. I play electric guitar. I like Pink Floyd as well!"

#convert this corpus to documents/sentences

documents_sentences=nltk.sent_tokenize(corpus_paragraph)
print("Type of documents_sentences :",type(documents_sentences))
print(documents_sentences)

#The above code directly works as this "sent_tokenize" function is direclty callable without creating any class


#Convert this corpus to vocabulary/words

words=nltk.word_tokenize(corpus_paragraph)
print("Type pf words variable :",type(words))
print(words)

#Keep fullstops and commas, etc. in separate index as well

#Treebank Tokenizer 

# Creating object treebank 
#treebank=nltk.TreebankWordTokenizer()

# Invoking object words_treebank
#words_treebank=treebank.tokenize(corpus_paragraph)

words_treebank=nltk.TreebankWordTokenizer().tokenize(corpus_paragraph)

print("Type of variable :",words_treebank)
print(words_treebank)

# It doesnt keep fullstops, commas, explanation marks, etc. in separate indexes
```
